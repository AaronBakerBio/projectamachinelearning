Now I will get a baseline of the following models, using 3 fold cross validation. I will not optimize hyperparams and just use defaults in this step

```python
r_state = 43
n_folds = 3
y_train_df = y_train_df.values.reshape(-1)

# Transform the reviews into a count matrix using CountVectorizer

# Logistic Regression
start = time.time()
logistic_values = train_models_and_calc_scores_with_bag_of_words(log_reg(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("Logistic Regression", logistic_values, n_folds, start, end)

# Ridge Classifier
start = time.time()
ridge_values = train_models_and_calc_scores_with_bag_of_words(ridge(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("Ridge Classifier", ridge_values, n_folds, start, end)

# k-Nearest Neighbors
start = time.time()
knn_values = train_models_and_calc_scores_with_bag_of_words(knn(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("k-Nearest Neighbors", knn_values, n_folds, start, end)

# Random Forest
start = time.time()
randomforest_values = train_models_and_calc_scores_with_bag_of_words(randomforest(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("Random Forest", randomforest_values, n_folds, start, end)

# Linear SVC
start = time.time()
linearsvc_values = train_models_and_calc_scores_with_bag_of_words(linearsvc(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("Linear SVC", linearsvc_values, n_folds, start, end)

# Log-loss SGD
start = time.time()
loglossSGD_values = train_models_and_calc_scores_with_bag_of_words(loglossSGD(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("log-loss SGD", loglossSGD_values, n_folds, start, end)

# Nearest Centroid
start = time.time()
nearestcentroid_values = train_models_and_calc_scores_with_bag_of_words(nearestcentroid(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("Nearest Centroid", nearestcentroid_values, n_folds, start, end)

# Complement Naive Bayes
start = time.time()
compnaivebayes_values = train_models_and_calc_scores_with_bag_of_words(compnaivebayes(), x_reviews, y_train_df, n_folds=n_folds, random_state=r_state)
end = time.time()
print_results("Complement Naive Bayes", compnaivebayes_values, n_folds, start, end)
```

Complement Naive Bayes had the highest AUROC among the models that could calculate it and also achieved the best validation accuracy. Although other models may have been faster, the main point we care about is the best AUROC.

```python
alpha_values = np.arange(.1, 5, 0.1)
fit_prior = [True, False]
norm = [True, False]
results = []
n_folds = 3
r_state = 43
y_train_df = y_train_df.values.reshape(-1)

for alpha in alpha_values:
    print(f"running for alpha {alpha}")
    for fit in fit_prior:
        for n in norm:
            cnb_model = compnaivebayes(alpha=alpha, fit_prior=fit, norm=n, force_alpha=True)
            cnb_values = train_models_and_calc_scores_with_bag_of_words(
                cnb_model, x_reviews, y_train_df, n_folds=n_folds, random_state=r_state
            )
            results.append((alpha, fit, n, cnb_values))

# Sort by best AUROC (which is cnb_values[4] based on the assumption that the fifth value is AUROC)
results.sort(key=lambda x: x[3][4], reverse=True)

# Print top 3 models based on AUROC
for i in range(3):
    alpha, fit, n, cnb_values = results[i]
    print(f"Rank {i+1}")
    print(f"Model: Complement Naive Bayes | alpha: {alpha}, fit_prior: {fit}, norm: {n}")
    print(f"rmse for train: {cnb_values[0]:.4f}, rmse for val: {cnb_values[1]:.4f}")
    print(f"accuracy for train: {cnb_values[2]:.4f}, accuracy for val: {cnb_values[3]:.4f}")
    print(f"AUROC for val: {cnb_values[4]:.4f}\n")
```

The only hyperparams we can modify in our case are alpha (smoothing), fit prior (using class priors (need to explain)) and normalizing term frequencies. Overall, I found that the highest accuracy was when using class priors and normalizing term frequency, with an alpha of .8

```python
ngram_range_options = [(1, 1), (1, 2), (1, 3)]  # N-gram ranges to test
n_folds = 3
random_state = 43
best_auroc_score = -1
best_model_values = None
best_ngram_range = None

# Iterate over the n-gram ranges
for ngram_range_option in ngram_range_options:
    print(f"\nTesting n-grams: {ngram_range_option}")

    # Call the train_with_ngram_and_tfidf_thresholding function to train the model
    # with the n-gram filtering and no TF-IDF thresholding
    best_ngram_range, complement_naive_bayes_values, average_auroc = train_with_ngram_and_tfidf_thresholding(
        compnaivebayes(alpha=0.8, fit_prior=True, norm=True),  # Model
        x_reviews,  # Input reviews
        y_train_df,  # Labels (fixed to directly pass the dataframe)
        ngram_ranges=[ngram_range_option],  # Testing single n-gram range at a time
        use_tfidf=False,  # No TF-IDF thresholding
        n_folds=n_folds,  # Cross-validation folds
        random_state=random_state  # Random state for reproducibility
    )

    print_results2(f"N-gram Model (ngram_range={ngram_range_option})", complement_naive_bayes_values, n_folds)

    # Check if this is the best AUROC
    if average_auroc > best_auroc_score:
        best_auroc_score = average_auroc
        best_model_values = complement_naive_bayes_values
        best_ngram_range = ngram_range_option

# Print the best AUROC and corresponding n-gram range
print(f"\nBest AUROC: {best_auroc_score:.4f} with n-gram range: {best_ngram_range}")
```

It appears ngrams 1,3 with stopwords thresholded at the ends are useful and improve our AUROC. I will not test full size ngrams without stopwords removed as it is incredibly slow and computationally expensive. This appears slower than before. However, we can improve our speed by reducing the bag size with tfidf. Our goal is to minimally reduce the AUROC by 1% with maximal tfidf threshold

```python
percentiles_to_test = range(1, 99, 1)

# Initialize parameters for the best result with TF-IDF thresholding
best_auroc_with_tfidf = 0.8843  # Starting AUROC to beat
best_tfidf_values = None
best_tfidf_percentile = None
auroc_threshold_drop = 0.01  # Stop if AUROC drops by 0.01 or more
number_of_folds = 3
random_state_value = 43

# Fixed n-gram range found from previous tests
ngram_range_to_use = (1, 3)

# Iterate over TF-IDF percentiles to test
for current_percentile in percentiles_to_test:
    print(f"\nTesting TF-IDF percentile: {current_percentile}%")

    # Call the train_with_ngram_and_tfidf_thresholding function to train the model
    # with the best n-gram range and applying TF-IDF thresholding at the current percentile
    best_ngram_range, tfidf_model_values, current_auroc = train_with_ngram_and_tfidf_thresholding(
        compnaivebayes(alpha=0.8, fit_prior=True, norm=True),  # Model
        x_reviews,  # Input reviews
        y_train_df,  # Labels
        ngram_ranges=[ngram_range_to_use],  # Fixed n-gram range
        use_tfidf=True,  # Enable TF-IDF thresholding
        tfidf_percentile=current_percentile,  # TF-IDF percentile threshold
        n_folds=number_of_folds,  # Number of folds for cross-validation
        random_state=random_state_value  # Random state for reproducibility
    )

    # Print the results for the current percentile
    print_results2(f"TF-IDF Percentile Threshold: {current_percentile}%", tfidf_model_values, number_of_folds)

    # Check if this is the best AUROC
    if current_auroc > best_auroc_with_tfidf:
        best_auroc_with_tfidf = current_auroc
        best_tfidf_values = tfidf_model_values
        best_tfidf_percentile = current_percentile
    elif current_auroc < best_auroc_with_tfidf - auroc_threshold_drop:
        # Stop if AUROC drops by 0.01 or more
        print(f"\nAUROC dropped from {best_auroc_with_tfidf:.4f} to {current_auroc:.4f}. Stopping early.")
        break

# Print the best AUROC and percentile
print(f"\nBest AUROC with TF-IDF filtering: {best_auroc_with_tfidf:.4f} at {best_tfidf_percentile}% TF-IDF threshold")
```

```python
best_ngram_range = (1, 3)
best_alpha = 0.8
best_fit_prior = True
best_norm = True
best_tfidf_percentile = 14

vectorizer = CountVectorizer(ngram_range=best_ngram_range)
x_reviews_counts = vectorizer.fit_transform(x_reviews)

tfidf_transformer = TfidfTransformer()
x_reviews_tfidf = tfidf_transformer.fit_transform(x_reviews_counts)
feature_names = vectorizer.get_feature_names_out()
avg_tfidf_scores = np.array(x_reviews_tfidf.mean(axis=0)).flatten()
tfidf_threshold_value = np.percentile(avg_tfidf_scores, best_tfidf_percentile)

words_to_keep = [feature_names[i] for i in range(len(feature_names)) if avg_tfidf_scores[i] > tfidf_threshold_value]
filtered_vectorizer = CountVectorizer(vocabulary=words_to_keep, ngram_range=best_ngram_range)

x_reviews_filtered = filtered_vectorizer.fit_transform(x_reviews)

final_model = compnaivebayes(alpha=best_alpha, fit_prior=best_fit_prior, norm=best_norm)
final_model.fit(x_reviews_filtered.toarray(), y_train_df)

x_test_df = pd.read_csv(os.path.join('data_reviews', 'x_test.csv'))
x_test_counts = filtered_vectorizer.transform(x_test_df['text'])
yprob_test = final_model.predict_proba(x_test_counts.toarray())[:, 1]

file_path = "yproba1_test.txt"
with open(file_path, 'w') as f:
    for prob in yprob_test:
        f.write(f"{prob}\n")

print(f"Predictions saved to {file_path}")
```

```python
import matplotlib.pyplot as plt

# Create a dictionary to store the results based on different conditions
results_dict = {}

# Collect results in a dictionary instead of a list for easier plotting
for alpha in alpha_values:
    print(alpha)
    for fit in fit_prior:
        for n in norm:
            cnb_model = compnaivebayes(alpha=alpha, fit_prior=fit, norm=n, force_alpha=True)
            cnb_values = train_models_and_calc_scores_with_bag_of_words(
                cnb_model, x_reviews, y_train_df, n_folds=n_folds, random_state=r_state
            )
            key = (fit, n)  # Key by fit_prior and norm
            if key not in results_dict:
                results_dict[key] = {'alpha': [], 'auroc': []}
            results_dict[key]['alpha'].append(alpha)
            results_dict[key]['auroc'].append(cnb_values[4])  # Assuming AUROC is the 5th value

# Plot the AUROC for each combination of fit_prior and norm
plt.figure(figsize=(10, 6))

for key, values in results_dict.items():
    fit, n = key
    label = f"fit_prior: {fit}, norm: {n}"
    plt.plot(values['alpha'], values['auroc'], label=label)

# Adding plot labels and title
plt.xlabel('Alpha Value')
plt.ylabel('AUROC')
plt.title('AUROC vs Alpha for Different fit_prior and norm Combinations')
plt.legend(title="Model Variants")
plt.grid(True)
plt.show()
```
